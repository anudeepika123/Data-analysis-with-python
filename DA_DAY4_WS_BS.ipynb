extracting a table from a webpage

```python
url="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
page1=requests.get(url)
code1=BeautifulSoup(page1.text,'html')
print(code1)
```

```python
t1=code1.find('table')
print(t1)
```

```python
h=t1.find_all('th')
print(h)
```

```python
Heading=[i.text for i in h]
print(Heading)
```

```python
import pandas as pd
mydf=pd.DataFrame(columns=Heading)
print(mydf)
```

```python

rows=t1.find_all('tr')
print(rows)
```

```python

rows1=t1.find_all('td')
print(rows1)
```

```python
data_act=[i.text for i in rows1]
print(data_act)
```

```python
df2=pd.DataFrame(columns=['Name','Rank','NW','Company'])
```

```python
ind=0
start=0
stop=4
for i in data_act:
  if stop<=40:
    mydf.loc[ind]=data_act[start:stop]
    start=start+4
    stop=stop+4
    ind=ind+1
print(mydf)

```

```python
mydf.to_csv("richest.csv")
pd.read_csv("/content/richest.csv")
```
