```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
```

```python
df2=pd.read_csv("/content/Project3 (2).csv")
df2
```

```python
feature1=['Income','LoanAmount','MonthsEmployed','DTIRatio']
feature2=df2.loc[0:99,feature1]#features
Target2=df2.loc[0:99,"Status"]
Target2
#target variable
```

```python
from sklearn.linear_model import LinearRegression
model2=DecisionTreeClassifier()
model2.fit(feature2,Target2)
```

```python
v1=int(input("enter Income"))
v2=int(input("enter LoanAmount"))
v3=input("enter MonthsEmployed")
v4=input("enter DTIRatio")
predict=model2.predict([[v1,v2,v3,v4]])
if predict == 1:
  print("Loan will be sanctioned")
else:
  print("Improve your credits and try again")
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from sklearn.ensemble import RandomForestClassifier
```

```python
model2=RandomForestClassifier(n_estimators=10)
#train the model
model2.fit(feature2,Target2)
pre2=model2.predict([[84208,129188,26,0.31]])
print(pre2)
if pre2 == 1:
  print("loan is sactioned")
else:
  print("loan is not sactioned")
```

```python
import os
output_dir1="tree_visual"
os.makedirs(output_dir1,exist_ok=True)
for i,tree in enumerate(model2.estimators_):
  tree_dot_file1=os.path.join(output_dir1,f"tree_{i}.dot")
  tree_png_file1=os.path.join(output_dir1,f"tree_{i}.png")
```

```python
model2.estimators_
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file1, feature_names=['Income','LoanAmount','MonthsEmployed','DTIRatio'],
                    class_names=[str(cls) for cls in model2.classes_], filled=True, rounded=True)
```

```python
command1 = f"dot -Tpng {tree_dot_file1} -o {tree_png_file1}"
os.system(command1)
print(f"Tree {i} visualization saved to {tree_png_file1}")
```
