```python
#creating set
s={1,2,3,4,'hi',"anu"}
print(s)
```

```python
#create set using functions
s1=set((1,"priya",'b'))
print(s1)
```

```python
#length of the set
s={"uma",123,'a'}
print(len(s))
```

```python
#union
s1={1,2,3}
s2={4,5,6}
print(s1|s2)
```

```python
#intersection
s1={1,2,4,5,3}
s2={4,5,6}
print(s1&s2)
```

```python
#difference
s1={1,2,4,5,3}
s2={4,5,6}
print(s1-s2)
```

```python
#checks whether subset or not
s1={1,2,4,5,3}
s2={4,5}
print(s2.issubset(s1))
```

```python
#delete element from the set
s1={1,44,85,9,2,4,5,3}
s1.discard(44)
print(s1)
```

```python
#converting tuple into set
t=set(("anu",'a',123))
print(t)
```

```python
#type of the set
s={"anu",'a',123}
print(type(s))
```

```python
#creating a tuple
t=(1,2,3,4)
print(t)
```

```python
#accesing elements
print(t[-1])

print(t[0])
```

```python
#creating tuple
t=tuple(("anu",'a',123))
print(t)
```

```python
#accesing using +indexes
t=(8,45,69,12,47,56,12)
print(t[:2])
print(t[1:])
```

```python
#concatnating two tuples
t=(12,3,4)
t1=(1,2,3)
print(t+t1)
```

```python
##accesing using -indexes
t=(8,45,69,12,47,56,12)
print(t[-1:])
print(t[:-2])
```

```python
#zipping of two elements
t=(1,2,4,5,3)
t1=(4,5,6,13,15)
a=zip(t,t1)
print(tuple(a))
```

```python
#length of tuple
t1=(1,2,3)
print(len(t1))
```

```python
#duplicating of tuples
t1=(1,2,3)
print(t1*2)
```

```python
#print the values in tuple

t=(8,45,69,12,47,56,12)
for i in t:
  print(i)
```
