```python
#there are 5 subjects,psychology,nano technology,medicine,onthology,and anthropology
#two students have applied to do masters in any of the 5 fields with line plot show in which field they can excel
#the cut off for each subject is 70.
#std1=[67,45,73,71,57],std2=[43,67,68,53,85]
std1=[67,45,73,71,57]
sub=["psychology","nanotechnology","medicine","onthology","anthropology"]#plt.plottype(X,y,color,marker,label,linewidth,linestyle)
plt.plot(sub,std1,label="std1",color="b",marker="P")
std2=[43,67,68,53,85]
plt.plot(sub,std2,label="std2",color='y',marker="H")
cutoff=[70,70,70,70,70]
plt.plot(cutoff,label="cutoff",color='r',linestyle=":")
plt.xlabel("subjects")
plt.ylabel("Marks")
plt.legend()

```

```python
#An India vs aus match is on live india has given 298 as target for aus.the runs for every 5 overs is given as follows:
#india=[25,51,84,131,160,189,220,250,267,297] for 5 overs
#aus =[15,41,94,110,151]
#plot overs vs runs for both the countries with india as blue line and aus as yellow line.give appropriate labels and legend.the marker for india is flag and marker for aus is plus
#take overs from arange funtn create runs as arrays.
import numpy as np
india=np.array([25,51,84,131,160,189,220,250,267,297])
aus=np.array([15,41,94,110,151])
overs_aus=np.arange(5,26,5)
overs_ind=np.arange(5,51,5)
plt.title("Ind vs Aus")
plt.plot(overs_ind,india,color="b",label="ind",marker=">",linewidth=4)
plt.plot(overs_aus,aus,color="y",label="aus",marker="P",linewidth=4)
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.grid()
```
